```python
#read the sets data
sets = pd.read_csv('data/sets.csv')
sets.head()
```

```python
#1 solution code below 
sets.sort_values('year').head()
```

```python
#2 solution code below 
sets[sets['year']==1949]
```

```python
#3 solution code below
sets.sort_values('num_parts', ascending=False).head()
```

```python
#visualize data 
#how does the number of sets released in 1955 compare to the number of sets released in 2019
sets_by_year=sets.groupby('year').count()
sets_by_year['set_num'].head()
```

```python
sets_by_year['set_num'].tail()
```

```python
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```

```python
#calcualte the number of different themes by calendar year
themes_by_year = sets.groupby('year').agg({'theme_id':pd.Series.nunique})
```

```python
themes_by_year.rename(columns = {'theme_id': 'nr_themes'}, inplace= True)
themes_by_year.head()
```

```python
themes_by_year.tail()
```

```python
#plot the number of themes released from 1949 to 2019
plt.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2])
```

```python
#compare number of themes as sets by using 2 seperate axes
ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color='g')
ax2.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], 'b')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of sets', color='green')
ax2.set_ylabel('Number of Themes', color='blue')
```

```python
#average number of parts per LEGO set
parts_per_set = sets.groupby('year').agg({'num_parts': pd.Series.mean})
```

```python
#visualize the data
plt.scatter(parts_per_set.index[:-2], parts_per_set.num_parts[:-2])
```

```python
#number of sets per LEGO theme which theme has the most 
themes = pd.read_csv('data/themes.csv')
themes.head()
```

```python
#merge data and create Bar Charts 
set_theme_count = sets['theme_id'].value_counts()
set_theme_count[:5]
```

```python
set_theme_count = pd.DataFrame({'id':set_theme_count.index,
                                'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id')
merged_df[:3]
```

```python
#plot the top ten themes on a chart
plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)
 
plt.bar(merged_df.name[:10], merged_df.set_count[:10])
```
